# NEO4J



Vous avez pour projet de modéliser un data center dans l'objectif d'évaluer l'analyse d'impacte d'une panne matérielle ou logicielle sur les services offerts aux utilisateurs.

Exemple : un serveur de base de données tombe en panne entrainant une indisponibilité du service de base de données lequel impacte toutes les applications web dont les sites de ventes de certains clients.

Le modéle Entité-Association suivant représente les relations qui lient les différentes entités métier.

Un hypveriseur est une machine physique qui instancie de 0 à N machines virtuelles. 

Une machine virtuelle peut être instanciée uniquement sur un et un seul hyperviseur.

Chaque machine virtuelle disposera d'un nom et d'une IP.

Une machine virtuelle sera utilisée pour déployer un ou plusieurs services informatiques et un service informatique sera disponible uniquement sur une seule machine virtuelle.

On notera qu'un service informatique peut s'appuyer sur d'autres services informatiques.

Par exemple, un site web sera considéré comme un service informatique lequel s'appuie sur une base de données pour le contenu et un service web (nginx) pour la publication des pages html.
