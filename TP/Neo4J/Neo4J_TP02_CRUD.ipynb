```{note} Objectifs

Le but de ce TP est de vous initier aux opérations de création, de lecture, de mise à jour et de suppression (CRUD) en utilisant le langage Cypher.

Au cours de ce TP, vous apprendrez à:

- **Créer des nœuds et des propriétés :** Vous apprendrez comment définir et ajouter de nouveaux nœuds dans la base de données, ainsi que comment attribuer et gérer les propriétés associées à ces nœuds.
- **Établir des relations :** Vous découvrirez comment lier les nœuds entre eux à travers divers types de relations, essentiels pour structurer le réseau de données.
- **Construire des sous-graphes :** Vous serez guidé dans la création de structures de données plus complexes, composées de plusieurs nœuds et relations, formant des sous-ensembles cohérents au sein de la base de données globale.
- **Explorer les relations à divers niveaux de profondeur :** Vous explorerez les capacités de Cypher pour naviguer à travers le réseau de nœuds à différentes profondeurs, permettant une analyse détaillée et flexible des connexions entre données.

```



# Pré-requis 

Démarrer votre container neo4j :

Vérifiez que votre container Neo4J est opérationnel :

Pour interagir avec le serveur Neo4J , vous utiliserez l'interface web native de Neo4J.
Cliquez sur le lien suivant :

http://127.0.0.1:7474

Vous êtes à présent prêt pour pratiquer le cypher.

Dans ce TP, vous apprendrez à créer et modifier les propriétés des noeuds et des relations.

Pour plus d'information sur la syntaxe du langage suivez le lien ci-dessous :

https://neo4j.com/docs/cypher-manual/3.5/





## Création d'un graphe

### 1 - Création des noeuds

Pour créer un noeud, vous devrez utiliser la commande CREATE. La documentation est disponible ici :

https://neo4j.com/docs/cypher-manual/current/clauses/create/

Pour créer un simple noeud avec un label et des propriétés :

L'option RETURN n'est pas obligatoire mais cela nous permet de voir le noeud créé. Un noeud peut avoir plusieurs label si besoin :

Vous pouvez créer plusieurs noeuds consécutivement en séparant la définition des noeuds par une ",".

### 2 - Création des relations

Il existe 3 maniéres d'établir les relations :

* Créer les noeuds en même temps que les relations
* Créer des relations sur des noeuds existants
* Créer les noeuds et les relations en fonction des noeuds / relations existante 
https://neo4j.com/docs/cypher-manual/current/clauses/merge/#query-merge-node-derived)


#### Créer les noeuds en même temps que les relations

On utilise la commande CREATE et on spécifie les relations qui lient les noeuds entre eux :
    


#### Créer des relations sur des noeuds existants

Les noeuds existent et vous voulez ajouter une relation. On définira avec un match le noeud de départ de la relation et le noeud de destination :




La relation "KNOWS" allant du noeud "Tormund" au noeud "Jaime" a été ajoutée.

#### Créer les noeuds et les relations en fonction des noeuds / relations existantes

La commmande merge permet de spécifier le comportement à appliquer si le pattern est trouvé ou non.

La commande suivante cherche le sous graphe `Daenerys-[LOVES]->Jon' :
 * si celui-ci n'est pas trouvé alors on initialise la propriété since de la relation LOVES au 01/03/2018
 * sinon on met à jour la propriété updated à aujourd'hui



Dans notre cas, le personnage Jon existe dans notre graphe mais comme Daenerys n'existe pas, Neo4j va créer le noeud Daenerys et la relation LOVES avec Jon.

# Exercice

Dans cette exercice, on vous demande de modéliser l'arbre généalogique suivant :

#### 1 - Créer l'arbre généalogique ci-dessus en créant les noeuds en même temps que les relations :

L'ensemble des noeuds disposeront d'un Label Cartoon.

## 2 - Vérifier votre modéle en fournissant les requêtes Cypher qui répondront aux questions suivantes :

####  Afficher uniquement le sous-graphe que vous venez de créer :

#### Donnez la descendance de Sangoku ?

#### Qui sont les parents de Sangohan ?

#### Indiquez les petits enfants de Hercule ?

#### Donnez la liste des couples ?

## 3 - Les opérations CRUD

#### Modifier le graphe de manière à ce que tous les enfants de Sangoku est une relation CONNAIT avec le noeud goku

#### Modifier le graphe de manière à ce que les personnages au moins grand-parents est une relation EST_ANCETRE_DE avec leur petits enfants 

#### Supprimez l'ensemble les relations de type EST_ANCETRE_DE :

#### Ajoutez l'année de naissance de Sangoku (1998) comme attribut ainsi que sa date de mariage 1995 :

#### Ajoutez l'attribut est grandparent à tous les personnages qui ont une descendance supérieur ou égale à 2 :

#### Supprimer l'attribut grandparent à tous les noeuds Cartoon :
