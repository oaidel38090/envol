# Présentation de Jupyter

Les notebooks Jupyter sont des cahiers électroniques qui, dans le même document, peuvent rassembler du texte, des images, des formules mathématiques et du code informatique exécutable. Ils sont manipulables interactivement dans un navigateur web.

Initialement développés pour les langages de programmation Julia, Python et R (d'où le nom Jupyter), les notebooks Jupyter supportent près de 40 langages différents.

La cellule est l'élément de base d'un notebook Jupyter. Elle peut contenir du texte formaté au format Markdown, du code brut "Raw" ou du code informatique exécutable.



Pour exécuter le contenu de cette cellule, vous avez plusieurs possibilités :

* Cliquer sur le menu Run, puis Run Selected Cells.



* Cliquer sur le bouton Run (sous la barre de menu).



* Presser simultanément les touches Ctrl + Entrée.


Pour identifier les cellules contenant du code exécutable, elles sont reconnaissables par des [].

````python
print("c'est du code exécutable appuyer sur touches Ctrl + Entrée pour voir le résultat")
````

Lorsque vous exécutez une cellule, les crochets [] se voit affecter un entier pour préciser l'ordre d'exécution de la cellule.

Cette information est pratique pour identifier l'ordre d'exécution d'une séquence de cellules.
Si vous exécutez une seconde fois la cellule, la valeur entre crochet sera incrémenter de un.

Toutes les cellules du notebook sont éditables, pour cela il vous suffit de cliquer 2 fois dessus pour qu'elles passent en mode éditable.

Vous pouvez ajouter autant de cellules que vous souhaitez avec l'icône + :
